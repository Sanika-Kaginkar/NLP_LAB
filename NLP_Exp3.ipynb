**1. Load the Dataset**

```python
import pandas as pd

# Load the dataset
file_path = 'path_to_your_dataset.csv'  # Replace with your actual file path
data = pd.read_csv('/content/words_pos.csv')

# Display the first few rows of the dataset
print(data.head(20))
```

**2. Data Preparation**

```python
# Convert the dataset into a list of sentences
sentences = []
current_sentence = []

for index, row in data.iterrows():
    word = row['word']
    tag = row['pos_tag']
    if word == '.':  # Assuming sentences are separated by periods
        sentences.append(current_sentence)
        current_sentence = []
    else:
        current_sentence.append((word, tag))

if current_sentence:
    sentences.append(current_sentence)

# Extract unique tags and words
tags = set(tag for sentence in sentences for word, tag in sentence)
words = set(word for sentence in sentences for word, tag in sentence)
```

**3. Calculate Transition Matrix**

```python
from collections import defaultdict

# Initialize transition matrix
transition_matrix = defaultdict(lambda: defaultdict(int))

# Count tag transitions
for sentence in sentences:
    for i in range(len(sentence) - 1):
        current_tag = sentence[i][1]
        next_tag = sentence[i + 1][1]
        transition_matrix[current_tag][next_tag] += 1

# Convert counts to probabilities
for current_tag in transition_matrix:
    total_transitions = sum(transition_matrix[current_tag].values())
    for next_tag in transition_matrix[current_tag]:
        transition_matrix[current_tag][next_tag] /= total_transitions

# Convert to DataFrame for better visualization
transition_df = pd.DataFrame(transition_matrix).fillna(0)
print("Transition Matrix:")
print(transition_df)
```

**4. Calculate Emission Matrix**

```python
# Initialize emission matrix
emission_matrix = defaultdict(lambda: defaultdict(int))

# Count word emissions
for sentence in sentences:
    for word, tag in sentence:
        emission_matrix[tag][word] += 1

# Convert counts to probabilities
for tag in emission_matrix:
    total_emissions = sum(emission_matrix[tag].values())
    for word in emission_matrix[tag]:
        emission_matrix[tag][word] /= total_emissions

# Convert to DataFrame for better visualization
emission_df = pd.DataFrame(emission_matrix).fillna(0)
print("Emission Matrix:")
print(emission_df)
```
